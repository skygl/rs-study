## 결합 확률 (Joint Probability)

동시에 고려해야 하는 사건에 대한 확률 (= 사건들이 동시에 일어날 확률)

사건 A와 B의 교집합의 확률을 계산하는 것과 같습니다.

$$P(A \cap B) = P(A,B)$$

예를 들어 트럼프 카드에서 카드를 뽑는 경우를 생각해볼 수 있습니다.

사건 A : "빨간색" 카드를 뽑는다.

- $P(A) = {26\over52} = 0.5$

사건 B : "숫자 6" 카드를 뽑는다.

- $P(B) = {4\over52} = 0.0769...$

이 때 사건 A와 사건 B의 결합 확률은 "빨간색 숫자 6" 카드를 뽑을 확률

- $P(A \cap B) = P(A,B) = {2\over52} = 0.0384...$

## 결합 확률 함수 (Joint Probability Function)

결합 확률을 함수로 표현한 것

## 결합 확률 질량 함수 (Joint Probability Mass Function)

두 이산 확률 변수 X, Y가 각각 $X=x_1,x_2,...,x_m$, $Y=y_1,y_2,...,y_n$을 취할 수 있을 때 X가 $x_i$의 값을 갖고, Y가 $y_i$의 값을 가질 확률은 다음과 같이 나타낼 수 있습니다.

$$P(X=x_i,Y=y_i)=f(x,y)(i=1,2,...,m,j=1,2,...,n)$$

여기서 $f(x,y)$을 이산 확률 변수 X,Y의 결합 확률 질량 함수(= 결합 확률 분포)라고 합니다. 확률 변수 X가 어떤 특정한 값을 가지고, 확률 변수 Y가 어떤 특정한 값을 가질 때의 확률을 나타내는 함수입니다.

### 성질

1. $f(x,y)\geq0$
2. $\Sigma_x\Sigma_yf(x,y)=1$
3. x, y 평면 상의 어떤 영역 A에 대해 $P[(x,y)\in A]=\Sigma\Sigma_Af(x,y)$이다.

### 예시

빨간 공 3개, 노란 공 2개, 초록 공 4개가 들어있는 주머니에서 임의로 2개의 공을 꺼낼 때 뽑은 빨간 공의 개수를 X라고 하고, 뽑은 노란 공의 개수를 Y라고 설정합니다.
확률 변수 X, Y가 취할 수 있는 값은 모두 0, 1, 2가 됩니다. 단 두 변수의 합이 2를 초과할 수는 없습니다.

## 결합 확률 밀도 함수 (Joint Probability Density Function)

두 개의 연속 확률 변수 X, Y가 존재할 때 X가 a 이상 b 이하면서 Y가 c이상 d이하일 확률은 다음과 같이 나타냅니다.

$$P(a \leq X \leq b, c \leq Y \leq d ) = \int_a^b \int_c^d f(x,y)dxdy$$

여기서 $f(x,y)$를 연속 확률 변수 X, Y의 결합 확률 밀도 함수라고 부릅니다.

### 성질

- 모든 $(x,y)$에 대하여 $f(x,y)\geq 0$
- $\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}f(x,y)dydx=1$
- $P[(x,y)\in A]=\int \int_A f(x,y) dydx$
